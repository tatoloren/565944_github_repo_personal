Las listas por comprensión permiten crear nuevas listas aplicando una expresión a cada elemento de un iterable, con sintaxis concisa y legible:

```python

```

## Parte 2 — Lambda con `map` y `filter`

Las funciones lambda son funciones anónimas de una sola expresión:

```python
lambda parámetro: expresión
```

Combinadas con `map` y `filter` permiten transformar y filtrar listas de forma funcional:

```python
list(map(lambda x: x * 2, lista))       # transforma
list(filter(lambda x: x > 0, lista))    # filtra
```

## Parte 3 — Lambda con `sorted` y `reduce`

`sorted` permite ordenar cualquier iterable usando una función clave:

```python
sorted(lista, key=lambda x: x["campo"], reverse=False)
```

`reduce` aplica una función acumulativa sobre una lista para obtener un único resultado:

```python
from functools import reduce
reduce(lambda acumulador, elemento: acumulador + elemento, lista, valor_inicial)
```
