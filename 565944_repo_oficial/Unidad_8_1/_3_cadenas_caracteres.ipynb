# 🔤 Cadenas de Caracteres (`str`) en Python — Guía Completa

Una **cadena** (o `string`) es una secuencia de caracteres entre comillas simples o dobles:

```python
s = "hola"
s = 'hola'   # equivalente
```

Las cadenas son **inmutables**: no se pueden modificar en el lugar. Los métodos siempre devuelven una **nueva cadena**.

En este notebook vemos:
1. Acceso y operaciones básicas
2. Métodos esenciales (transformación)
3. Búsqueda: `find`, `rfind`, `startswith`, `endswith`
4. Alineación y formato visual
5. Validación: métodos `is...`
6. Conversión: `split`, `partition`, `join`
7. Formato de cadenas: `%`, `.format()`, f-strings


---
## Parte 1: Acceso y Operaciones Básicas

Las cadenas se comportan como secuencias: se puede acceder por índice, hacer slicing, concatenar y repetir, igual que con las listas.

````python
s = "hello"
#    01234   <- indices

# Acceso por indice
print(f"s[0]:    '{s[0]}'")   # 'h'
print(f"s[1]:    '{s[1]}'")   # 'e'
print(f"s[-1]:   '{s[-1]}'")  # 'o'  (ultimo)

# Slicing
print(f"s[1:4]:  '{s[1:4]}'")  # 'ell'
print(f"s[::-1]: '{s[::-1]}'") # 'olleh' (invertida)
````

````python
s1 = "hello"
s2 = "world"

# Concatenacion con +
print(f"s1 + s2:          '{s1 + s2}'")          # 'helloworld'
print(f"s1 + ' ' + s2:    '{s1 + ' ' + s2}'")    # 'hello world'

# Repeticion con *
print(f"s1 * 3:           '{s1 * 3}'")            # 'hellohellohello'

# Longitud
print(f"len(s1):          {len(s1)}")              # 5
````

> ⚠️ **Las cadenas son inmutables.** No se puede modificar un carácter directamente:
> ```python
> s = "hello"
> s[0] = "H"   # ❌ TypeError: 'str' object does not support item assignment
> ```
> Para "modificar" una cadena, hay que crear una nueva.

## Parte 2: Métodos Esenciales

Todos estos métodos devuelven una **nueva cadena** sin modificar la original.

````python
texto = "hOLA, mUnDo"

print(f"upper():      '{texto.upper()}'")       # 'HOLA, MUNDO'
print(f"lower():      '{texto.lower()}'")       # 'hola, mundo'
print(f"capitalize(): '{texto.capitalize()}'")  # 'Hola, mundo' (solo 1ra letra)
print(f"title():      '{texto.title()}'")       # 'Hola, Mundo' (cada palabra)
print(f"swapcase():   '{texto.swapcase()}'")    # invierte mayus/minus
````

````python
# strip() elimina espacios (u otros caracteres) de ambos extremos
s = "   hello   "
print(f"strip():  '{s.strip()}'")    # 'hello'
print(f"lstrip(): '{s.lstrip()}'")   # 'hello   '  (solo izquierda)
print(f"rstrip(): '{s.rstrip()}'")   # '   hello'  (solo derecha)

# Tambien elimina caracteres especificos
print(f"strip('-'): '{'---hola---'.strip('-')}'")  # 'hola'
````

````python
s = "hello"

# replace(viejo, nuevo) reemplaza TODAS las ocurrencias
print(f"replace('l','r'):    '{s.replace('l', 'r')}'")    # 'herro'

# Con un tercer argumento, limita cuantas veces reemplaza
print(f"replace('l','r',1):  '{s.replace('l', 'r', 1)}'") # 'herlo'
````

````python
s = "Hola, mundo"

# find() -> indice de la primera ocurrencia (-1 si no existe)
print(f"find('mundo'):  {s.find('mundo')}")   # 6
print(f"find('xyz'):    {s.find('xyz')}")     # -1

# rfind() -> busca desde el final
s2 = "Hola, mundo, mundo"
print(f"rfind('mundo'): {s2.rfind('mundo')}")  # 13

# startswith() / endswith() -> verifica extremos de la cadena
print(f"startswith('Hola'):  {s.startswith('Hola')}")   # True
print(f"endswith('mundo'):   {s.endswith('mundo')}")    # True
print(f"startswith('mundo'): {s.startswith('mundo')}")  # False
````

````python
# split(sep, maxsplit) -> divide en lista
cadena = "uno,dos,tres,cuatro,cinco"
print(f"split(','):    {cadena.split(',')}")        # ['uno','dos','tres','cuatro','cinco']
print(f"split(',', 2): {cadena.split(',', 2)}")     # ['uno','dos','tres,cuatro,cinco']

# rsplit(sep, maxsplit) -> divide desde el final
print(f"rsplit(',', 2): {cadena.rsplit(',', 2)}")    # ['uno,dos,tres','cuatro','cinco']

# splitlines() -> divide por saltos de linea
ml = "linea uno\nlinea dos\nlinea tres"
print(f"splitlines():  {ml.splitlines()}")           # ['linea uno','linea dos','linea tres']

# partition(sep) -> devuelve tupla (antes, sep, despues) en la 1ra ocurrencia
s = "manzana pera manzana uva"
print(f"partition:  {s.partition('manzana')}")      # ('', 'manzana', ' pera manzana uva')
print(f"rpartition: {s.rpartition('manzana')}")     # ('manzana pera ', 'manzana', ' uva')

# join() -> une lista en cadena con un separador
palabras = ['Hola', 'mundo', 'Python']
print(f"' '.join():  {' '.join(palabras)}")          # 'Hola mundo Python'
print(f"'-'.join():  {'-'.join(palabras)}")          # 'Hola-mundo-Python'
````

## Parte 3: Alineación y Formato

Estos métodos son útiles para dar formato visual a textos, como alinear columnas o rellenar con ceros.

````python
s = "hola"

# center() centra el texto en un ancho dado, rellenando con un caracter
print(f"center(10,'-'): '{s.center(10, '-')}'")  # '---hola---'

# ljust() alinea a la izquierda
print(f"ljust(10,'-'):  '{s.ljust(10, '-')}'")   # 'hola------'

# rjust() alinea a la derecha
print(f"rjust(10,'-'):  '{s.rjust(10, '-')}'")   # '------hola'

# zfill() rellena con ceros a la izquierda (util para numeros)
print(f"'42'.zfill(6):  '{'42'.zfill(6)}'")      # '000042'
````

````python
# Ejemplo practico: imprimir una tabla alineada
productos = [("Manzana", 1.50), ("Banana", 0.80), ("Cereza", 3.20)]

print("Producto".ljust(12) + "Precio".rjust(8))
print("-" * 20)
for nombre, precio in productos:
    print(nombre.ljust(12) + f"${precio:.2f}".rjust(8))
````

## Parte 4: Métodos de Validación (`is...`)

Estos métodos devuelven `True` o `False` según el contenido de la cadena. Son muy útiles para validar entradas del usuario.

````python
# isalpha() -> solo letras
print(f"'Python'.isalpha():    {'Python'.isalpha()}")   # True
print(f"'Python3'.isalpha():   {'Python3'.isalpha()}")  # False

# isalnum() -> letras y/o numeros
print(f"'Python3'.isalnum():   {'Python3'.isalnum()}")  # True
print(f"'+++'.isalnum():       {'+++'.isalnum()}")       # False

# isdigit() / isnumeric() / isdecimal() -> solo numeros
print(f"'12345'.isdigit():     {'12345'.isdigit()}")     # True
print(f"'-12345'.isdigit():    {'-12345'.isdigit()}")    # False (el signo no es digito)
print(f"'123.45'.isdecimal():  {'123.45'.isdecimal()}")  # False (el punto no es decimal)

# isspace() -> solo espacios en blanco
print(f"'   '.isspace():       {'   '.isspace()}")       # True
print(f"'  _  '.isspace():     {'  _  '.isspace()}")     # False

# islower() / isupper() -> minusculas / mayusculas
print(f"'python'.islower():    {'python'.islower()}")    # True
print(f"'Python'.islower():    {'Python'.islower()}")    # False
print(f"'PYTHON'.isupper():    {'PYTHON'.isupper()}")    # True
print(f"'Python'.isupper():    {'Python'.isupper()}")    # False

# istitle() -> cada palabra empieza con mayuscula
print(f"'Python Es Genial'.istitle(): {'Python Es Genial'.istitle()}")  # True
print(f"'Python es genial'.istitle(): {'Python es genial'.istitle()}")  # False
````

### Diferencia entre `isdigit()`, `isnumeric()` y `isdecimal()`

Para la mayoría de los casos prácticos son equivalentes, pero difieren en caracteres especiales:

| Método | `'123'` | `'²'` (superíndice) | `'½'` (fracción) |
|---|---|---|---|
| `isdecimal()` | ✅ | ❌ | ❌ |
| `isdigit()` | ✅ | ✅ | ❌ |
| `isnumeric()` | ✅ | ✅ | ✅ |

> 💡 Para validar si el usuario ingresó un número entero común, usá `isdecimal()`.

## Parte 5: Operador `in` / `not in`

Permite verificar si una subcadena está contenida dentro de otra. Devuelve `True` o `False`.

````python
s = "hola mundo"

print(f"'hola' in s:      {'hola' in s}")        # True
print(f"'mundo' in s:     {'mundo' in s}")       # True
print(f"'adios' in s:     {'adios' in s}")       # False
print(f"'adios' not in s: {'adios' not in s}")   # True

# Uso practico: validar email
email = "usuario@ejemplo.com"
if '@' in email and '.' in email:
    print("El email parece valido")
else:
    print("El email no parece valido")
````

## Parte 6: Formato de Cadenas

Python ofrece tres formas de insertar variables dentro de un string. La más moderna y recomendada son las **f-strings**.

| Estilo | Sintaxis | Versión |
|---|---|---|
| Operador `%` | `"texto %s" % var` | Clásico (Python 2) |
| `.format()` | `"texto {}".format(var)` | Python 3+ |
| f-string | `f"texto {var}"` | Python 3.6+ ✅ recomendado |


````python
legajo = 12345
nombre = "Maria"
nota = 9.5

# Estilo antiguo con %
print("Legajo: %d, Nombre: %s, Nota: %.1f" % (legajo, nombre, nota))

# .format() con posicion o nombre
print("Legajo: {}, Nombre: {}, Nota: {}".format(legajo, nombre, nota))
print("Legajo: {0}, Nombre: {1}, Nota: {2}".format(legajo, nombre, nota))

# f-string (recomendado): permite expresiones directamente
print(f"Legajo: {legajo}, Nombre: {nombre}, Nota: {nota}")
print(f"Nombre en mayusculas: {nombre.upper()}")
print(f"Nota aprobada: {nota >= 6}")

# Formato numerico con f-strings
precio = 1234.5678
print(f"Precio normal:     {precio}")
print(f"2 decimales:       {precio:.2f}")
print(f"Separador de miles:{precio:,.2f}")
print(f"Relleno con ceros: {legajo:08d}")
````

## 📊 Resumen Completo de Métodos

### Transformación
| Método | ¿Qué hace? |
|---|---|
| `upper()` / `lower()` | Todo en mayúsculas / minúsculas |
| `capitalize()` | Primera letra en mayúscula, el resto en minúscula |
| `title()` | Primera letra de cada palabra en mayúscula |
| `swapcase()` | Invierte mayúsculas y minúsculas |
| `strip()` / `lstrip()` / `rstrip()` | Elimina caracteres de los extremos |
| `replace(v, n)` | Reemplaza `v` por `n` |
| `zfill(n)` / `center()` / `ljust()` / `rjust()` | Relleno y alineación |

### Búsqueda
| Método | ¿Qué hace? |
|---|---|
| `find(s)` / `rfind(s)` | Índice de `s` (desde el inicio / desde el final) |
| `startswith(s)` / `endswith(s)` | Verifica si empieza / termina con `s` |
| `in` / `not in` | Verifica si una subcadena está contenida |

### Conversión
| Método | ¿Qué hace? |
|---|---|
| `split(sep)` / `rsplit(sep)` | Divide en lista (desde inicio / final) |
| `splitlines()` | Divide por saltos de línea |
| `partition(sep)` / `rpartition(sep)` | Devuelve tupla `(antes, sep, después)` |
| `sep.join(lista)` | Une lista en cadena con `sep` |

### Verificación
| Método | Devuelve `True` si... |
|---|---|
| `isalpha()` | Solo letras |
| `isdigit()` / `isnumeric()` / `isdecimal()` | Solo números (con variantes) |
| `isalnum()` | Solo letras y/o números |
| `isspace()` | Solo espacios en blanco |
| `islower()` / `isupper()` | Toda en minúsculas / mayúsculas |
| `istitle()` | Cada palabra empieza con mayúscula |
