## PARTE 1 — Listas por Comprensión

Una **lista por comprensión** nos permite crear una nueva lista aplicando una expresión a cada elemento de otra lista.

La estructura básica es:

```python
[expresion for elemento in lista]
```

Y con filtrado:

```python
[expresion for elemento in lista if condicion]
```

### 1.1 Transformación — Redondear hacia arriba

Queremos obtener una nueva lista con cada calificación redondeada hacia arriba (usando `math.ceil`).

> 💡 `math.ceil(3.5)` devuelve `4` — siempre redondea al entero superior.

````python
# Importar librerias
import math
````

````python
# Con for in (básico)
redondeadas = []
for nota in calificaciones:
    redondeadas.append(math.ceil(nota))

print("Forma tradicional: ", redondeadas)
````

````python
# Con lista por comprensión
redondeadas = [math.ceil(nota) for nota in calificaciones]

print("Lista original:", calificaciones)
print("Lista por comprensión:", redondeadas)
````

> ✅ Ambas versiones producen el mismo resultado. La lista por comprensión es más concisa y legible.

### 1.2 Filtrado — Calificaciones aprobadas

Ahora queremos obtener solo las calificaciones mayores o iguales a 4 (aprobadas).

````python
# # Con for in (básico)
aprobadas = []
for nota in redondeadas:
    if nota >= 4:
        aprobadas.append(nota)

print("Forma tradicional: ", aprobadas)
````

````python
# Con lista por comprensión
aprobadas = [nota for nota in redondeadas if nota >= 4]

print("Lista por comprensión:", aprobadas)
````

## PARTE 2 — Lista de Diccionarios

Ahora vamos a enriquecer nuestra lista. En lugar de solo guardar la nota, vamos a representar cada estudiante como un **diccionario** con dos claves: `nombre` y `nota`.

````python
estudiantes = [
    {"nombre": "Ana",     "nota": 8},
    {"nombre": "Bruno",   "nota": 3.5},
    {"nombre": "Carla",   "nota": 9},
    {"nombre": "Diego",   "nota": 9.75},
    {"nombre": "Elena",   "nota": 3},
    {"nombre": "Franco",  "nota": 6.5}
]

print("Lista de estudiantes:")
for e in estudiantes:
    print(f"  {e['nombre']}: {e['nota']}")
````

### 2.1 Transformación sobre diccionarios — Redondear la nota

Creamos una nueva lista de diccionarios con la nota redondeada hacia arriba.

> 💡 Notar que la expresión ahora construye un nuevo diccionario `{...}` para cada elemento.

````python
import math
````

````python
estudiantes_redondeados = [{"nota": math.ceil(e["nota"])} for e in estudiantes ]

print("Notas redondeadas:")
for e in estudiantes_redondeados:
    print(f"  {e['nota']}")
````

### 2.2 Filtrado sobre diccionarios — Estudiantes aprobados

````python
aprobados = [e for e in estudiantes if e["nota"] >= 4]

print("Estudiantes aprobados:")
for e in aprobados:
    print(f"  {e['nombre']}: {e['nota']}")
````

## PARTE 3 — Funciones Lambda con `map` y `filter`

Una **función lambda** es una función anónima de una sola línea:

```python
lambda parametro: expresion
```

Se usa frecuentemente con:
- `map(funcion, lista)` → aplica la función a cada elemento (transformación)
- `filter(funcion, lista)` → filtra los elementos donde la función devuelve `True`

> ⚠️ Tanto `map` como `filter` devuelven un objeto especial. Hay que convertirlo con `list()` para verlo como lista.

````python
import math

# Definamos una función que redondee un número
def redondear_tradicional(numero):
    return math.ceil(numero)
````

````python
# Convertimos la función tradicional en una lambda
redondear_lambda = lambda numero: math.ceil(numero)
````

````python
print(f"Redondeo con método tradicional {redondear_tradicional(9.5)} y con lambda {redondear_lambda(9.5)}")
````

### 3.1 `map` + `lambda` — Redondear calificaciones

````python
print(calificaciones)
````

````python
redondeado_map = list(map( lambda nota  : math.ceil(nota)  , calificaciones ))
print(redondeado_map)
````

````python
# map aplica la lambda a cada elemento de la lista
redondeadas = list(map(lambda nota: math.ceil(nota), calificaciones))

print("Original:        ", calificaciones)
print("Con map+lambda:  ", redondeadas)
````

### 3.2 `filter` + `lambda` — Calificaciones aprobadas

````python
print(calificaciones)
````

````python
aprobados = list(filter(lambda nota: nota >=4 , calificaciones))
print(aprobados)
````

````python
# filter conserva solo los elementos donde la lambda devuelve True
aprobadas = list(filter(lambda nota: nota >= 4, redondeadas))

print("Original:         ", redondeadas)
print("Con filter+lambda:", aprobadas)
````

## PARTE 4 — Lambda con `sorted` y `reduce`

Las lambdas también son muy útiles para **ordenar** listas y para **reducir** una lista a un único valor.

### 4.1 `sorted` + `lambda` — Ordenar calificaciones

Podemos ordenar la lista de estudiantes por nota, de mayor a menor.

````python
# Ordenar la lista simple de menor a mayor
ordenadas_asc = sorted(calificaciones)
print("Ascendente: ", ordenadas_asc)

# Ordenar de mayor a menor
ordenadas_desc = sorted(calificaciones, reverse=True)
print("Descendente:", ordenadas_desc)

````

````python
print(calificaciones)
calificaciones_sorted = sorted(calificaciones,reverse=True)
print(calificaciones_sorted)
````

````python
print(estudiantes)
estudiantes_sorted = sorted(estudiantes, key= lambda e : e["nombre"], reverse=False)
print(estudiantes_sorted[:3])
````

````python
# Ordenar la lista de diccionarios por nota (de mayor a menor)
# La lambda le dice a sorted por qué campo ordenar
estudiantes_ordenados = sorted(estudiantes, key=lambda e: e["nota"], reverse=True)

print("\nEstudiantes ordenados por nota (mayor a menor):")
for e in estudiantes_ordenados:
    print(f"  {e['nombre']}: {e['nota']}")
````

### 4.2 `reduce` + `lambda` — Promedio de calificaciones

`reduce` aplica una función acumuladora sobre toda la lista, de izquierda a derecha, hasta obtener un único valor.

> 💡 `reduce` está en el módulo `functools` y hay que importarlo.

````python
print(calificaciones)
````

````python
from functools import reduce
suma_notas = reduce( lambda  acumulador , nota :  acumulador + nota  , calificaciones, 0)
print(suma_notas)
````

````python
print(estudiantes)
````

````python
from functools import reduce

# Primero sumamos todas las notas con reduce
# La lambda recibe el acumulador y el elemento actual, y devuelve su suma
suma_total = reduce(lambda acumulador, nota: acumulador + nota, calificaciones)

promedio = suma_total / len(calificaciones)

print("Calificaciones:", calificaciones)
print("Suma total:    ", suma_total)
print(f"Promedio:       {promedio:.2f}")
````

````python
print(estudiantes)
````

````python
from functools import reduce

# Ajustamos el código para usar la lista de diccionarios estudiantes en lugar de calificaciones
suma_total = reduce(lambda acumulador, e: acumulador + e["nota"], estudiantes, 0)

promedio = suma_total / len(estudiantes)

print("Suma total:    ", suma_total)
print(f"Promedio:       {promedio:.2f}")
````

---
## 🧠 Resumen comparativo

| Herramienta | ¿Qué hace? | Ejemplo |
|---|---|---|
| Lista por comprensión | Transforma o filtra una lista | `[ceil(x) for x in lista]` |
| `map` + `lambda` | Transforma cada elemento | `map(lambda x: ceil(x), lista)` |
| `filter` + `lambda` | Filtra según condición | `filter(lambda x: x >= 4, lista)` |
| `sorted` + `lambda` | Ordena por criterio | `sorted(lista, key=lambda e: e["nota"])` |
| `reduce` + `lambda` | Acumula en un solo valor | `reduce(lambda acc, x: acc + x, lista)` |
