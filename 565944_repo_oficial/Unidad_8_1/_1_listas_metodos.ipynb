# 📋 Listas en Python

Las **listas** son una de las estructuras de datos más importantes en Python. Permiten almacenar múltiples valores en una sola variable, en orden, y son **mutables** (se pueden modificar después de crearlas).

```python
mi_lista = [1, 2, 3, 'hola', True]
```

1. **Métodos** propios de las listas
2. **Funciones** built-in útiles para trabajar con listas
3. **Acceso, slicing y operaciones** básicas

---
## Parte 1: Métodos de las Listas

Los métodos son funciones que pertenecen a un objeto. Se invocan con la sintaxis `lista.metodo()`.

> 💡 A diferencia de las funciones, los métodos **modifican la lista original** (salvo excepciones como `index()` y `count()`).

### `append()` — Agregar un elemento al final

Agrega **un solo elemento** al final de la lista. La lista se modifica en el lugar.

````python
lst = [1, 2, 3]
lst.append(4)
print(f"append(): {lst}")  # [1, 2, 3, 4]
````

### `extend()` — Agregar varios elementos al final

A diferencia de `append()`, `extend()` toma un **iterable** (como otra lista) y agrega cada uno de sus elementos al final.

````python
lst = [1, 2]
lst.extend([3, 4])
print(f"extend(): {lst}")  # [1, 2, 3, 4]

# ⚠️ Diferencia clave: append vs extend
lst_a = [1, 2]
lst_b = [1, 2]
lst_a.append([3, 4])   # agrega la lista como UN elemento
lst_b.extend([3, 4])   # agrega los elementos de la lista
print(f"append([3,4]): {lst_a}")  # [1, 2, [3, 4]]
print(f"extend([3,4]): {lst_b}")  # [1, 2, 3, 4]
````

### `insert()` — Insertar en una posición específica

Inserta un elemento en el **índice indicado**, desplazando el resto hacia la derecha.

```python
lista.insert(índice, valor)
```

````python
lst = [1, 2, 3]
lst.insert(2, 1.5)  # en el índice 1 inserta 1.5
print(f"insert(): {lst}")  # [1, 1.5, 2, 3]
````

### `remove()` — Eliminar por valor

Elimina la **primera ocurrencia** del valor indicado. Si el valor no existe, lanza un `ValueError`.

````python
lst = [1, 2, 2, 3]
lst.remove(2)  # elimina solo el PRIMER "2"
print(f"remove(): {lst}")  # [1, 2, 3]
````

### `pop()` — Eliminar y devolver un elemento

`pop()` elimina un elemento y lo **devuelve** para que puedas usarlo.
- Sin argumento: elimina el **último** elemento.
- Con índice: elimina el elemento en esa posición.

````python
lst = [1, 2, 3]
ultimo = lst.pop()       # saca el último
print(f"pop() devuelve: {ultimo}")  # 3
print(f"Lista resultante: {lst}")   # [1, 2]

# Con índice
lst = [10, 20, 30, 40]
elemento = lst.pop(1)    # saca el índice 1 (valor 20)
print(f"pop(1) devuelve: {elemento}")  # 20
print(f"Lista resultante: {lst}")      # [10, 30, 40]
````

### `index()` — Buscar la posición de un valor

Devuelve el **índice** de la primera aparición del valor. Si no existe, lanza `ValueError`.

````python
lst = [1, 2, 3]
pos = lst.index(2)
print(f"index(2): {pos}")  # 1  → el "2" está en la posición 1
````

### `count()` — Contar ocurrencias

Cuenta cuántas veces aparece un valor en la lista.

````python
lst = [1, 2, 2, 3]
cantidad = lst.count(2)
print(f"count(2): {cantidad}")  # 2  → hay dos "2" en la lista
````

### `sort()` — Ordenar la lista

Ordena la lista **en el lugar** (modifica el original) en orden ascendente por defecto.

- `sort()` → orden ascendente
- `sort(reverse=True)` → orden descendente

````python
lst = [3, 1, 4, 1, 5, 9, 2]
lst.sort()
print(f"sort() ascendente:  {lst}")  # [1, 1, 2, 3, 4, 5, 9]

lst.sort(reverse=True)
print(f"sort() descendente: {lst}")  # [9, 5, 4, 3, 2, 1, 1]
````

### `reverse()` — Invertir el orden

Invierte el orden de los elementos de la lista (también modifica el original).

> ⚠️ `reverse()` **no ordena**, solo da vuelta el orden actual.

````python
lst = [1, 2, 3]
lst.reverse()
print(f"reverse(): {lst}")  # [3, 2, 1]
````

### 📊 Resumen de Métodos

| Método | ¿Qué hace? | ¿Modifica la lista? | ¿Devuelve algo? |
|---|---|---|---|
| `append(x)` | Agrega `x` al final | ✅ | ❌ |
| `extend(iter)` | Agrega elementos de otro iterable | ✅ | ❌ |
| `insert(i, x)` | Inserta `x` en el índice `i` | ✅ | ❌ |
| `remove(x)` | Elimina la primera ocurrencia de `x` | ✅ | ❌ |
| `pop(i)` | Elimina y devuelve el elemento en `i` | ✅ | ✅ |
| `index(x)` | Devuelve el índice de `x` | ❌ | ✅ |
| `count(x)` | Cuenta cuántas veces aparece `x` | ❌ | ✅ |
| `sort()` | Ordena en el lugar | ✅ | ❌ |
| `reverse()` | Invierte el orden en el lugar | ✅ | ❌ |

## Parte 2: Funciones Built-in para Listas

Python incluye funciones globales (no son métodos de lista) muy útiles para trabajar con colecciones.

````python
numeros = [4, 7, 2, 9, 1]

print(f"len():  {len(numeros)}")   # cantidad de elementos → 5
print(f"sum():  {sum(numeros)}")   # suma total           → 23
print(f"min():  {min(numeros)}")   # valor mínimo         → 1
print(f"max():  {max(numeros)}")   # valor máximo         → 9

# list() convierte otro iterable en lista
tupla = (1, 2, 3)
print(f"list(): {list(tupla)}")    # (1, 2, 3) → [1, 2, 3]
````

## Parte 3: Acceso y Operaciones

### Acceso por índice

Los índices en Python **empiezan en 0**. También se pueden usar índices negativos para contar desde el final.

````python
lst = [10, 20, 30, 40, 50]

print(f"lst[0]:  {lst[0]}")   # primer elemento  → 10
print(f"lst[2]:  {lst[2]}")   # tercer elemento  → 30
print(f"lst[-1]: {lst[-1]}")  # último elemento  → 50
print(f"lst[-2]: {lst[-2]}")  # penúltimo        → 40
````

### Slicing (rebanado)

Permite extraer una **sublista** usando la sintaxis `lista[inicio:fin:paso]`.
- El índice `inicio` está **incluido**.
- El índice `fin` está **excluido**.

````python
lst = [0, 1, 2, 3, 4, 5]

print(f"lst[1:4]:  {lst[1:4]}")   # del índice 1 al 3  → [1, 2, 3]
print(f"lst[:3]:   {lst[:3]}")    # desde el inicio    → [0, 1, 2]
print(f"lst[3:]:   {lst[3:]}")    # hasta el final     → [3, 4, 5]
print(f"lst[::2]:  {lst[::2]}")   # de a 2 pasos       → [0, 2, 4]
print(f"lst[::-1]: {lst[::-1]}")  # invertida          → [5, 4, 3, 2, 1, 0]
````

### Concatenación y Copia

````python
# Concatenación con +
lst1 = [1, 2]
lst2 = [3, 4]
print(f"lst1 + lst2: {lst1 + lst2}")  # [1, 2, 3, 4]

# Repetición con *
lst = [0] * 4
print(f"[0] * 4:     {lst}")          # [0, 0, 0, 0]

# Copia superficial con .copy()
original = [1, 2, 3]
copia = original.copy()
copia.append(99)
print(f"original: {original}")  # [1, 2, 3]  → no se modifica
print(f"copia:    {copia}")      # [1, 2, 3, 99]
````

### Desempaquetado (Unpacking)

Permite asignar los elementos de una lista a variables individuales en una sola línea.

````python
# Desempaquetado básico
a, b, c = [1, 2, 3]
print(f"a = {a}, b = {b}, c = {c}")  # a = 1, b = 2, c = 3

# Desempaquetado con * (captura el resto)
primero, *resto = [10, 20, 30, 40, 50]
print(f"primero: {primero}")  # 10
print(f"resto:   {resto}")    # [20, 30, 40, 50]
````
