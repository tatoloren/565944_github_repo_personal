# Listas

Una lista es una estructura de datos que nos permite almacenar múltiples valores dentro de una misma variable.

* es un arreglo de elementos
* es una estructura ordenada
* es mutable (puede cambiar)

```python
# Ejemplo de lista
numeros = [1, 2, 3, 4, 5]
```

## Características claves

1. **Ordenadas**: los elementos tienen una posición definida

```python
colores = ["rojo", "verde", "azul"]
```

2. **Indexadas**: se accede a sus elementos a través de un índice

```python
print(colores[0])  # rojo
print(colores[2])  # azul
```

* El índice empieza en 0
* Último elemento: len(lista) - 1

```python
print(f"El primer color es: {colores[0]}")  # El primer color es: rojo
print(f"El ultimo color es: {colores[-1]}")  # El ultimo color es: azul
```

3. **Mutables**: se pueden modificar.

A través de la función id() podemos validar que la posición de memoria se mantiene una vez modificada

```python
print(f"El id de colores es: {id(colores)}") 
colores.append("amarillo")
print(f"La nueva lista de colores es: {colores}")
print(f"El id de colores es: {id(colores)}")
```

4. **Heterogéneas**: admiten cualquier tipo de variable

```python
datos = ["Juan", 25, True]
print(datos)
```

## Operaciones básicas de listas

1. Agregar elementos, para ello usamos el método append

```python
# A la lista original le insertamos un nuevo elemento al final
numeros = [1, 2, 3, 4, 5]
numeros.append(6)
print(f"Lisna numeros actualizada: {numeros}")
```

2. Recorrer listas, para ello podemos usar for in

```python
# Usamos un bucle for para recorrer la lista numeros e imprimir cada elemento
for numero in numeros:
    print(numero)
```

```python
# Podemos agregar la función enumerate() para obtener el índice de cada elemento
for indice, numero in enumerate(numeros):
    print(f"Índice: {indice}, Número: {numero}")
```

## Listas como estructuras de datos (aplicación práctica)

Hasta ahora usamos listas como variables…
pero en realidad son nuestra primera estructura de datos en memoria

```python
notas = [7, 8, 6, 9]
```

Esto ya representa:

* un conjunto de datos
* una “tabla simple” (de un solo registro)
* algo que luego podría persistirse

```python
# Calculamos el promedio de las notas
suma = sum(notas)
promedio = suma / len(notas)
print(f"El promedio de las notas es: {promedio}")
```

¿Qué pasa si ahora necesitamos guardar múltiples listas relacionadas?

```python
# Generamos una lista para las notas de cada estudiante
notas_est_1 = [7, 8, 6, 9]
notas_est_2 = [7, 9, 6, 10]
notas_est_3 = [7, 6, 6, 8]
```

```python
# O podríamos ser mas eficientes y generar una lista de listas para almacenar las notas de cada estudiante
notas = [notas_est_1, notas_est_2, notas_est_3]
print(f"Las notas de los estudiantes son: {notas}")
```

# Uso de range

range() es una función de Python que genera una secuencia de números enteros.

No devuelve una lista directamente, sino un objeto iterable (pero se usa como si fuera una secuencia).

range() acepta uno o mas argumentos:

* si solo recibe un argumento, ese será el stop. Si ese argumento es 5, genera números entre 0 y 4
* si pasamos los tres argumentos, entonces el primero es el start, el segundo el stop y el tercero el step, por ejemplo range(1,10,2) genera 1, 3, 5, 7, 9

```python
# range con un solo argumento
secuencia = range(5) # 5 es el stop
print(f"Secuencia con un solo argumento: {secuencia}")
```

```python
# Convertimos la secuencia a una lista para mostrar los números
lista = list(secuencia)
print(f"Números en la secuencia: {lista}")
```

```python
# range con mas argumentos
secuencia = range(1, 10, 2) # 1 es el start, 10 es el stop, 2 es el step  
print(f"Números en la secuencia: {list(secuencia)}")
```
